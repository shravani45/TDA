## Distribution of transaction amount and transaction time

```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))
amount_val = df['Amount'].values
time_val = df['Time'].values
sns.histplot(amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
sns.histplot(time_val, ax=ax[1], color='b')
ax[0].set_xlim([min(amount_val), 500])
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
plt.show()
```

## finding and plotting outliers

```python
def find_outliers_iqr(data):
    outliers_count = {}
    for column in data.columns[:-1]:  # Exclude the target variable 'Class'
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        column_outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)][column]
        outliers_count[column] = len(column_outliers)
    return outliers_count
outliers_count_dict = find_outliers_iqr(df)

for column, count in outliers_count_dict.items():
    print(f"Number of outliers in {column}: {count}")
```

```python
num_columns = len(df.columns) - 1  # Exclude the target variable 'Class'
num_rows = (num_columns // 4) + 1  # Adjust the number of rows as needed

plt.figure(figsize=(16, 4 * num_rows))

# Loop through each column (excluding the target variable 'Class')
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(num_rows, 4, i+1)
    sns.boxplot(x='Class', y=column, data=df, palette='coolwarm')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```
